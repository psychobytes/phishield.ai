# Data Preparation

### dataset :
http://data.phishtank.com/data/online-valid.csv.gz [phishing url]
https://www.kaggle.com/datasets/siddharthkumar25/malicious-and-benign-urls [benign url]

```python
import pandas as pd

# load dataset
df_phish = pd.read_csv('phishtank.csv')
df_benign = pd.read_csv('kaggle_benign.csv')
#df_phish
#df_benign
```

```python
# create phishing dataframe
phishing = df_phish[['url']].sample(4000)
phishing['label'] = 1

# create benign dataframe
benign_filter = df_benign[df_benign['result'] == 0].head(4000)
benign = benign_filter[['url', 'result']]
benign['url'] = benign['url'].apply(lambda x: x if x.endswith('/') else x + '/')
benign.rename(columns={'result': 'label'}, inplace=True)
```

```python
# save dataframe to csv file
phishing.to_csv('phishing.csv', index=False)
benign.to_csv('benign.csv', index=False)
```

### Data Cleaning

```python
# load extracted data
df_pe = pd.read_csv('phishing_output.csv')
df_be = pd.read_csv('benign_output.csv')
```

```python
# check null data
df_pe.isna().sum()
df_be.isna().sum()
```

```python
# drop null data
df_pe = df_pe.dropna()
df_be = df_be.dropna()
```

```python
# change float to int
pe_float = ['feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8', 'feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13', 'feature_14', 'feature_15', 'feature_16', 'feature_17', 'feature_18', 'feature_19', 'feature_20', 'feature_21']
df_pe[pe_float] = df_pe[pe_float].astype(int)

be_float = ['feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8', 'feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13', 'feature_14', 'feature_15', 'feature_16', 'feature_17', 'feature_18', 'feature_19', 'feature_20', 'feature_21']
df_be[pe_float] = df_be[be_float].astype(int)
```

```python
# take only 1000 rows of data
df_pe = df_pe.sample(1000)
df_be = df_be.sample(1000)
```

```python
# change column name
df_pe.rename(columns={'feature_1':'have_ip', 'feature_2':'url_length', 'feature_3':'domain_length', 'feature_4':'special_char', 'feature_5':'punycode', 'feature_6':'shortener', 'feature_7':'url_slash', 'feature_8':'prefsuffix', 'feature_9':'subdomain', 'feature_10':'favicon', 'feature_11':'nonstandardport', 'feature_12':'tls', 'feature_13':'at', 'feature_14':'abnormalreq', 'feature_15':'anchor', 'feature_16':'link', 'feature_17':'mailto', 'feature_18':'forwarding', 'feature_19':'mouseover', 'feature_20':'right_click', 'feature_21':'popup'}, inplace=True)
df_be.rename(columns={'feature_1':'have_ip', 'feature_2':'url_length', 'feature_3':'domain_length', 'feature_4':'special_char', 'feature_5':'punycode', 'feature_6':'shortener', 'feature_7':'url_slash', 'feature_8':'prefsuffix', 'feature_9':'subdomain', 'feature_10':'favicon', 'feature_11':'nonstandardport', 'feature_12':'tls', 'feature_13':'at', 'feature_14':'abnormalreq', 'feature_15':'anchor', 'feature_16':'link', 'feature_17':'mailto', 'feature_18':'forwarding', 'feature_19':'mouseover', 'feature_20':'right_click', 'feature_21':'popup'}, inplace=True)
```
